```python
df1 = pd.read_csv("Questionário de avaliação de atividade prática - Ética profissional_2025_2 (respostas) - Respostas ao formulário 1 (3).csv")
df2 = pd.read_csv("Questionário de avaliação de atividade prática - Ética profissional (respostas) - Respostas ao formulário 1 (4).csv")
```

```python
df = pd.concat([df1, df2], ignore_index=True)
```

```python
print("Total de respostas:", len(df))
```

```python
if "Carimbo de data/hora" in df.columns:
    df = df.drop(columns=["Carimbo de data/hora"])
```

```python
col_eng = "Como você avaliaria seu grau de engajamento durante o desenvolvimento do projeto?"

df[col_eng].value_counts().plot(kind="bar")

plt.title("Engajamento dos estudantes")
plt.ylabel("Frequência")
plt.show()
```

```python
motivos = [
"Impacto na vida das pessoas",
"Impacto para várias pessoas",
"Criação de valor",
"Abordagem de um problema social",
"Interesse pessoal no tema",
"Interesse da equipe no tema",
"Grande potencial de discussão ética",
"Tema pouco explorado em jogos"
]
```

```python
for col in motivos:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df[motivos].mean().sort_values().plot(kind="barh")

plt.title("Importância média dos motivos")
plt.show()
```

```python
col_dif = "Quais foram as maiores dificuldades encontradas no desenvolvimento do jogo?"

dif = df[col_dif].dropna().str.split(", ").explode()

dif.value_counts().plot(kind="bar")

plt.title("Principais dificuldades")
plt.show()
```

```python
etapas = [
"Etapa 1 – Pesquisa e resumo do tema",
"Etapa 2 – Identificação das TAGs",
"Etapa 3 – Escolha do gênero do jogo",
"Etapa 4 – Criação do universo narrativo (lore, espaço, personagens)",
"Etapa 5 – Criação das regras e protótipo do jogo",
"Etapa 6 – Testes e ajustes",
"Etapa 7 – Produção da apresentação (slides e exposição)",
"Etapa 8 – Apresentação final"
]
```

```python
df[etapas].mean().sort_values().plot(kind="barh")

plt.title("Dificuldade média das etapas")
plt.show()
```

```python
def categorizar(texto):

    texto = str(texto).lower()

    if "ética" in texto or "ético" in texto:
        return "Aprendizado ético"

    if "equipe" in texto or "grupo" in texto:
        return "Trabalho em equipe"

    if "criatividade," in texto or "criar" in texto:
        return "Criatividade"

    if "jogo" in texto or "mecânica" in texto:
        return "Desenvolvimento de jogos"

    return "Outros"
```

```python
df["Categoria_aprendizado"] = df[col_aprendizado].apply(categorizar)
```

```python
def categorizar_aprendizado(texto):

    texto = str(texto).lower()

    # trabalho em equipe
    if "equipe" in texto or "grupo" in texto or "comunicação" in texto:
        return "Trabalho em equipe"

    # criação de jogos
    if "jogo" in texto or "tabuleiro" in texto or "regras" in texto:
        return "Criação de jogos"

    # criatividade
    if "criativ" in texto:
        return "Criatividade"

    # planejamento / organização
    if "tempo" in texto or "projet" in texto or "planej" in texto:
        return "Planejamento do projeto"

    # aprendizado sobre o tema
    if "tema" in texto or "conteúdo" in texto or "aprendi" in texto:
        return "Aprendizado sobre o tema"

    # diversão / experiência
    if "divert" in texto or "legal" in texto:
        return "Experiência positiva"

    return "Outros"
```

```python
df["Categoria_aprendizado"] = df[col_aprendizado].apply(categorizar_aprendizado)
```

```python
df["Categoria_aprendizado"].value_counts()
```

```python
sns.countplot(
    data=df,
    x="Categoria_aprendizado",
    order=df["Categoria_aprendizado"].value_counts().index
)

plt.title("Principais aprendizados relatados pelos estudantes")
plt.xlabel("Categoria")
plt.ylabel("Frequência")
plt.xticks(rotation=30)
plt.show()
```
